```python
def clean_Currency(x):
    if isinstance(x,str):
        return(x.replace('*','').replace(',',''))
    return(x)
```

```python
cricket_df['HS']=cricket_df['HS'].apply(clean_Currency).astype('int')
cricket_df
```

```python
cricket_df.Runs=cricket_df.Runs.astype('int')
cricket_df.Inns=cricket_df.Inns.astype('int')
cricket_df.NO=cricket_df.NO.astype('int')
cricket_df.HS=cricket_df.HS.astype('int')
cricket_df.Ave=cricket_df.Ave.astype('float')
cricket_df.BF=cricket_df.BF.astype('int')
cricket_df.SR=cricket_df.SR.astype('float')
cricket_df['100']=cricket_df['100'].astype('int')
cricket_df['50']=cricket_df['50'].astype('int')
cricket_df['0']=cricket_df['0'].astype('int')
cricket_df.dtypes
```
