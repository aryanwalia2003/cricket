```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from googleapiclient.discovery import build
import cufflinks as cf
import chart_studio.plotly as py
import plotly.express as px
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
init_notebook_mode(connected=True)
cf.go_offline()
import plotly.graph_objects as go
import json
```

```python
df=pd.read_html('https://myfinal11.in/series/vijay-hazare-trophy-122323/records/batting-most-runs/')
df[0].to_csv('batting_vht.csv')
batting_df=pd.read_csv('batting_vht.csv')
batting_df.head(20)
```

```python

index_names = batting_df[ batting_df['Player'] =='Player' ].index
batting_df=batting_df.drop(index_names)
```

```python
batting_df.replace(to_replace='-',value=0,inplace=True)
batting_df
```

```python
convert_dict = {'MAT': int,
                'INNS': int,
                'NO':int,
                'RUNS':int,
                'HS':int,
                'AVG':float,
                'BF':int,
                'SR':float,
                '100s':int,
                '50s':int,
                '6s':int,
                '4s':int,
                }
batting_df=batting_df.astype(convert_dict)
batting_df.dtypes
```

```python
batting_df[['sr.no','Name']] =batting_df['Player'].str.split(".",expand=True,)
batting_df.drop(columns='sr.no',inplace=True)
batting_df
```

```python
batting_df.drop(columns='Player',inplace=True)
```
